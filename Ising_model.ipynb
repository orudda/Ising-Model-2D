```python
import numpy as np
from numpy.random import rand
import matplotlib.pyplot as plt
from scipy import constants
```

# **Funções primordiais**

```python
def set_viz(viz,n,l):#troquem o nome dessa funçao
    for k in range(1,n+1,1):
        #vizinho direita
        if(k % l == 0):
            viz[k][1] = k + 1 - l
        else:
            viz[k][1] = k + 1
        #vizinho acima
        if(k > (n - l)):
            viz[k][2] = k + l - n
        else:
            viz[k][2] = k + l
        #vizinho esquerda 
        if((k-1) % l == 0):
            viz[k][3] = k + l - 1
        else:
            viz[k][3] = k - 1
        #vizinho baixo
        if(k < (l + 1)):
            viz[k][4] = k + n - l
        else:
            viz[k][4] = k - l
            
def gray_flip(t,n):
    k=t[0]
    if (k>n):
        return k,t
    t[k-1] = t[k]
    t[k] = k+1
    if (k != 1):
        t[0] = 1
    return k,t
   
def enumera_ising(n,s,viz):
    for i in range(len(s)):
        s[i]=-1
    g = np.zeros(4*n)
    t = np.arange(1,n+2)
    e = -(2*n)
    g[e] = -2
    lista=[]
    for i in range(1,2**n,1):
        k,t = gray_flip(t,n)
        h = s[int(viz[k][1])] + s[int(viz[k][2])] + s[int(viz[k][3])] + s[int(viz[k][4])] 
#         print(h)
        e = int(e + 2*s[k]*h)
        lista.append(e)
        g[e] = g[e] + 2
        s[k] = -s[k]
#         print(s[k])
    return g,lista
        
def media_termodinamica(g,lista,beta,n):
    z=0
    e_i = 0
    e_ii = 0
    eMin = min(lista)
    for e in set(sorted(lista)):
        if(g[e]<=0):
            continue
        e_1 = e - eMin
        z = z + (g[e]*np.exp(-beta*e_1))
        e_i = e_i + (e_1*g[e]*np.exp(-beta*e_1))
        e_ii = e_ii + ((e_1**2)*g[e]*np.exp(-beta*e_1))
    e_i = e_i/z
    e_ii = e_ii/z
    z = z*np.exp(-beta*eMin)
    cv = (beta**2) * ((e_ii - (e_i**2)))/n
    media = (e_i + eMin)/n
    return z,media,cv

def show(title, xlabel, ylabel, x_2, y_2, x_4, y_4, x_6, y_6):
  plt.plot(x_2, y_2, label = '2x2', color='blue')
  plt.plot(x_4, y_4, label = '4x4', color='green')
  plt.plot(x_6, y_6, label = '6x6', color='black')
  plt.title(title)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)
  plt.legend()
  plt.show()
```

```python
def questao_i(lista_2,lista_4,n_2,n_4):
  x_2=[]
  y_2=[]
  x_4=[]
  y_4=[]
  # x_6=[]
  # y_6=[]

  for e in lista_2:
    if (g_2[e] > 0):
      x_2.append(e/n_2)
      y_2.append(np.log(g_2[e]))
 
  for e in lista_4:
    if (g_4[e] > 0):
      x_4.append(e/n_4)
      y_4.append(np.log(g_4[e]))

  # for e in lista_6:
  #   if (g_6[e] > 0):
  #     x_6.append(e/n_6)
  #     y_6.append(np.log(g_6[e]))

  plt.plot(x_2, y_2, label = '2x2', color='blue')
  plt.plot(x_4, y_4, label = '4x4', color='green')
  # plt.plot(x_6, y_6, label = '6x6', color='black')
  plt.title('ln(g(E)) em função de E/N')
  plt.xlabel('E/N')
  plt.ylabel('ln(g(E))')
  plt.legend()
  plt.show()

    
def questao_ii_iii(t,g_2,g_4,g_6,lista_2,lista_4,lista_6,flag):
  x_2=[]
  y_2=[]
  x_4=[]
  y_4=[]
  x_6=[]
  y_6=[]
  for i in t:
      beta=1/i
      z_2,media_2,cv_2 = media_termodinamica(g_2,lista_2,beta,n_2)
      x_2.append(i)
      if (flag==2):
        y_2.append(media_2)
      else:
        y_2.append(cv_2)


      z_4,media_4,cv_4 = media_termodinamica(g_4,lista_4,beta,n_4)
      x_4.append(i)
      if (flag==2):
        y_4.append(media_4)
      else:
        y_4.append(cv_4)

      z_6,media_6,cv_6 = media_termodinamica(g_6,lista_6,beta,n_6)
      x_6.append(i)
      if (flag==2):
        y_6.append(media_6)
      else:
        y_6.append(cv_6)
  
  if (flag==2):
        show('energia por <e> em funcao da temperatura', 'T (Temperatura)', '<e>(Energia por Spin)', x_2, y_2, x_4, y_4, x_6, y_6)
  else:
    show('calor especifico em funcao da temperatura', 'T (Temperatura)', 'cv(Calor Especifico)', x_2, y_2, x_4, y_4, x_6, y_6)
  

def questao_iv_v(t,n_2,n_4,n_6,g_2,g_4,g_6,lista_2,lista_4,lista_6,flag):
  x_2=[]
  y_2=[]
  x_4=[]
  y_4=[]
  x_6=[]
  y_6=[]
  for i in t:
      beta=1/i
      z_2,media_2,cv_2 = media_termodinamica(g_2,lista_2,beta,n_2)
      x_2.append(i)
      aux_2 = -((1/n_2)*i*np.log(z_2))
      if (flag==4):
        y_2.append(aux_2) 
      else:
        y_2.append((media_2-aux_2)/i)    


      z_4,media_4,cv_4 = media_termodinamica(g_4,lista_4,beta,n_4)
      x_4.append(i)
      aux_4 = -((1/n_4)*i*np.log(z_4))
      if (flag==4):
        y_4.append(aux_4) 
      else:
        y_4.append((media_4-aux_4)/i)  

      z_6,media_6,cv_6 = media_termodinamica(g_6,lista_6,beta,n_6)
      x_6.append(i)
      aux_6 = -((1/n_6)*i*np.log(z_6))
      if (flag==4):
        y_6.append(aux_6) 
      else:
        y_6.append((media_6-aux_6)/i)  

  if (flag==4):
        show('energia livre em funcao da temperatura', 'T (Temperatura)', 'f(t)(Energia livre)', x_2, y_2, x_4, y_4, x_6, y_6)
  else:
    show('entropia em funcao da temperatura', 'T (Temperatura)', 's(t)(Entropia por spin)', x_2, y_2, x_4, y_4, x_6, y_6)
```
